```python
#i certainly think we can drop this features as it has less impact on model 
as_data.drop(["Country","Ad Topic Line","City","Timestamp"],inplace=True,axis=1)
```

```python
as_data.head()
```

```python
#independent and dependent
X=as_data.drop("Clicked on Ad",axis =1)
y=as_data["Clicked on Ad"]
```

```python
#spliting train and testing
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python

from sklearn.linear_model import LogisticRegression
```

```python
#our model
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
#predicting
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test,predictions)
```
